```python
exp_oper = pd.read_excel("Data\\5.Расходные операции_янв-дек2021.xlsx")
```

```python
exp_oper = exp_oper.drop(['Тип операции', 'Источник операции', 'Описание позиции', 'Описание организации источника', 'Общий объем', 'ЕИ объема'], axis=1)
```

```python
def Code_replace(x):
    x = x.replace("304", 'СЗ')
    x = x.replace("702", 'У')
    x = x.replace("101", 'М')
    x = x.replace("К04", 'М')
    x = x.replace("К05", 'М')
    x = x.replace("107", 'М')
    x = x.replace("402", 'М')
    return x
```

```python
exp_oper['Код организации источника'] = exp_oper['Код организации источника'].astype(str)
```

```python
exp_oper['Код организации источника'].unique()
```

```python
exp_oper['Код организации источника'] = exp_oper['Код организации источника'].apply(Code_replace)
```

```python
exp_oper['№ Площадки'].unique().shape
```

```python
def replace_num_site(x):
    x = x.replace("PL_", '')
    x = x.replace("КР_", '')
    x = x.replace("КР", '')
    x = x.replace(" (300)", '')
    x = x.replace(" (400)", '')
    x = x.replace(" (900)", '')
    x = x.replace(" (200)", '')
    x = x.replace(" (700)", '')
    x = x.replace(" (400) (400)", '')
    x = x.replace(" (К00)", '')
    return x
```

```python
exp_oper['№ Площадки'] = exp_oper['№ Площадки'].astype(str).apply(replace_num_site)
```

```python
exp_oper["Количество операции"] = exp_oper["Количество операции"].apply(lambda x: x*-1)
```

```python
exp_oper
```

```python
exp_oper.columns = ["Date", "item_ID", "Stock", "Site", "Cnt", "Mes", "Vol"]
```

```python
exp_oper.to_csv("Data//exp_oper.csv")
```
