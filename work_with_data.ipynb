```python
import pandas as pd
import numpy as np
```

```python
forNS = pd.read_csv("Data\\forNS.csv")
forNS['NS'] = 1
```

```python
forNS
```

```python
forSWAP = pd.read_csv("Data\\forSWAP.csv")
forSWAP['SWAP'] = 1
```

```python
forSWAP
```

```python
data_plan = pd.concat([forSWAP, forNS])
```

```python
data_plan = data_plan.rename(columns={'Date':'Date_plan'})
```

```python
data_plan
```

```python
data_plan = data_plan.fillna(0).drop(["Unnamed: 0"], axis = 1)
```

```python
data_plan = data_plan.sort_values(by=['Date_plan'])
```

```python
expOper = pd.read_csv("Data\\exp_oper.csv")
```

```python
expOper = expOper.drop(['Unnamed: 0', 'Vol'], axis=1)
```

```python
expOper = expOper.rename(columns={'Date':'Date_Fact'})
```

```python
expOper = expOper.sort_values(by=['Date_Fact'])
```

```python
data = expOper.merge(data_plan, on=['Site', 'Stock']).drop(['Mes'],axis=1)
```

```python
data
```

```python
data['Date_Fact'] = pd.to_datetime(data['Date_Fact'])
data['Date_plan'] = pd.to_datetime(data['Date_plan'])
```

```python
data['difference'] = data['Date_Fact'] - data['Date_plan']
```

```python
data
```

```python
d = pd.read_csv("Data\\Delivery.csv")
```

```python
d
```

```python
d = d.drop(['Unnamed: 0'], axis=1)
```

```python
data['difference'] = data['difference'].astype(str).apply(lambda x: x.replace(' days','')).astype(int)
```
